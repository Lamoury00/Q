```python
#Longest substring without repeated
#Given a string s, find the length of the longest substring without repeating characters.
#s = "abcabcbb"
#idea: give lp and rp, two pointer, iterate, rp move until repeated char
#stored len of lp to rp
#lp move to rp location, and rp move again
#so on and so forth

class Solution:
    def lengthOfLongestSubstring(self,s:str) -> int:
        n = len(s) #input string size limit the loop times
        ans = 0 #default 0 , finally output
        mp = {} # mp stores the current index of a character, {a:2,b:3} means char a has index 2, char b has index 3
        lp = 0

        for rp in range(n): # try to extend the range [i, j]
            if s[rp] in mp: # check jth position char is in the mp dict or not
                lp = max(mp[s[rp]],lp)
                #if we checked the char in mp dict, we update i to jth index number
                #use max because i might be larger than index number in mp dict, so i stayed
            ans = max(ans,rp-lp+1) #rp - lp +1 represent the length of the string
            mp[s[rp]] = rp+1 #update dict index
        
        return ans
```

```python
#String to Integer (atoi)
#convert string into a 32-bits integer
#rules:
# 输入字符串为空、或其他不合法情况，返回0；
# 字符串开头的空格要在预处理中删掉；
# 处理可能出现的正负号“+”，“-”，正负号只能出现一次；
# 超出整数范围的值取整数范围的边界值。


#idea:
#determine str or not str, not return 0
#strip lead whitespace and tail whitespace
#notice edge case that input is whitespace str like "         ", after strip is a '' empty string, watch out the index later on

class Solution:
    def myAtoi(self, s: str) -> int:

        if not s:
            return 0
        
        s = s.strip()
        #edge case, if input str is a str with empty spcace, "     "
        #after s.strip, still a '' string
        
        if s == '':
            return 0
        
        number = 0 #final output
        flag = 1 #represent positive or negative number
        
        if s[0] == '-':
            s = s[1:]
            flag = -1
        elif s[0] == '+':
            s = s[1:]
        
        for c in s:
            if c >= '0' and c <= '9':
                number = 10*number + ord(c) - ord('0')  #use unicode to represent a number
            else:
                break
        
        number = flag * number
        
        num_max =pow(2,31)-1  #limits
        num_min =-pow(2,31)   #limits
        
        if number <= num_max:
            number = number
        else:
            number = num_max
        
        if number >= num_min:
            number = number
        else:
            number = num_min
            
        return number
```

```python
#roman to integer
#create a values dict for roman to integer


# Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

# I can be placed before V (5) and X (10) to make 4 and 9. 
# X can be placed before L (50) and C (100) to make 40 and 90. 
# C can be placed before D (500) and M (1000) to make 400 and 900.
# Given a roman numeral, convert it to an integer.

# According to romann number rule, we have to consider substraction as edge case

values = {
    "I":1,
    "V":5,
    "X":10,
    "L":50,
    "C":100,
    "D":500,
    "M":1000,
}


class Solution:
    def romanToInt(self, s: str) -> int:
        total = 0  #final output
        lp = 0  #pointer
        
        while lp < len(s):
            #substraction
            if lp +1 < len(s) and values[s[lp]] < values[s[lp+1]]: #pointer is still in range, value less than next position value
                total += values[s[lp+1]] - values[s[lp]]
                lp += 2  #move 2 postision
                
            #not substraction
            else:
                total += values[s[lp]]
                lp += 1
            
        return total

```

```python
#3Sum

# Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.
# Notice that the solution set must not contain duplicate triplets.

# 在给定的数组中判断是否存在三个数的和是0，返回所有的组合，但是返回的组合中不能有重复。

from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = set() #final output, use set to remove duplicates
        dups = set() #hashset, store duplicate in order to skip next time
        visit = {}  #{value:index}
        
        for i , val1 in enumerate(nums):  #get index, value from nums
            if val1 not in dups: #if the value not in dups, not duplicated
                dups.add(val1) #add into duplicate hashset
                
                for j , val2 in enumerate(nums[i+1:]): #after add to dups set, then get index and values from string after index i
                    complement = -val1 -val2 #complement would be negative value of val1 and val2, ensure adding together is 0
                    
                    if complement in visit and visit[complement] == i: #value in visit dict and index in visit dict
                        res.add(tuple(sorted((val1,val2,complement)))) # sort,make a tuple and add to res hashset
                    visit[val2] = i #update visit, val2 and index
        
        return res
```

```python
#Remove duplicate from sorted array

#Input: nums = [0,0,1,1,1,2,2,3,3,4]
#Output: 5, nums = [0,1,2,3,4,_,_,_,_,_]

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        lp = 0

        while lp < len(nums)-1:
            if nums[lp] == nums[lp+1]:
                del nums[lp]
            else:
                lp += 1
        
        return len(nums)
```

```python
#Next Permutation

#idea: two pointers, reverse order
#from back to front, find the position when nums[i] < nums[i+1]
#find the number nums[k] > nums[i], then switch
#rest of , behand i, from i+1 to end, do sort
#only need next, so the next then break

class Solution:
    def nextPermutation(self, arr: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        if len(arr) < 1:
            return
        
        for i in range(len(arr)-2,-1,-1):
            if arr[i] < arr[i+1]:
                for k in range(len(arr)-1,-i,-1):
                    if arr[k] > arr[i]:
                        arr[i], arr[k] = arr[k], arr[i]
                        arr[i+1:] = sorted(arr[i+1:])
                        break
                break
            else:
                if i == 0:
                    return arr.sort()
```

```python
# Multiply string
# Given two non-negative integers num1 and num2 represented as strings, return the product of num1 and num2, also represented as a string.
# output would be string

class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        if num1 == "0" or num2 =="0":
            return "0"
        
        #we need to know result number of digits, using list[0] to represent
        N = len(num1) + len(num2)
        answer = [0]*N  #[0,0,0,0,0] number of 0 depends on N
        
        #reverse
        first_number = num1[::-1]
        second_number = num2[::-1]
        
        for place2, digit2 in enumerate(second_number):
            for place1, digit1 in enumerate(first_number):
                cur_pos = place1 + place2
                
                carry = answer[cur_pos] #before start, need to get last step carry
                
                mul = int(digit1) * int(digit2) + carry
                
                answer[cur_pos] = mul % 10  #last digit of mul, write into cur_pos
                answer[cur_pos + 1] += mul // 10 #write into next position,for next loop carry
                
        if answer[-1] == 0:  #remove zeros
            answer.pop()
            
        return ''.join( str(x) for x in reversed(answer)) # cannot convert list directly to string, has to do every element in list converting to string and concatate all togather
```

```python
#group anagrams
#anagrams means word or phases using same letters but rearranging into different words and phases

# Input: strs = ["eat","tea","tan","ate","nat","bat"]
# Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

from collections import defaultdict
from typing import List

class Solution:
    def groupAnagrams(self, input):

        ans = defaultdict(list) #create a hashmap, in python is dict, then stores numbers of letters as patterns in keys, and strings in values
        
        for s in input:
            count = [0] * 26  #each give a 26 0s
        
            for c in s:
                count[ord(c)-ord('a')] +=1 #find the position from 'a',then +=1
        
            ans[tuple(count)].append(s) #tupling same patterns, appending diff strings
        
        return ans.values() #output all strings
        
        
```

```python
#Add binary
#Given two binary strings a and b, return their sum as a binary string.

# Example 1:

# Input: a = "11", b = "1"
# Output: "100"
# Example 2:

# Input: a = "1010", b = "1011"
# Output: "10101"

class Solution:
    def addBinary(self,a:str,b:str) -> str:
        n = max(len(a),len(b))  #find longer length n

        a=a.zfill(n)
        b=b.zfill(n) #making two the same length, using zfill to fill 0 until same length

        carry = 0
        ans = []

        for i in range(n-1,-1,-1):
            if a[i] == '1':
                carry +=1
            if b[i] == '1':
                carry +=1
            
            #这里用mod,二进制相加有几种情况，11,10,01,00，如果是11和00余数为2，10和01是余数为1
            if carry % 2 == 1:
                ans.append('1')
            else:
                ans.append('0')
            carry //= 2  # keeping 整除2，取整数
        
        if carry == 1:
            ans.append('1')
        
        ans.reverse()
        return ''.join(ans)

```

```python
# Minimum Window Substring


# Given two strings s and t of lengths m and n respectively, 
# return the minimum window substring of s such that every character in t (including duplicates) is included in the window. 
# If there is no such substring, return the empty string "".

# The testcases will be generated such that the answer is unique.

# A substring is a contiguous sequence of characters within the string.

# Example 1:

# Input: s = "ADOBECODEBANC", t = "ABC"
# Output: "BANC"

# Example 2:

# Input: s = "a", t = "a"
# Output: "a"

# Example 3:

# Input: s = "a", t = "aa"
# Output: ""

from collections import Counter

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if not s or not t:
            return ""
        
        dict_t = Counter(t)  #{'char': # of occurence}
        required = len(dict_t) #how many char we have in t
        filtered_s = []
        
        for i , char in enumerate(s):
            if char in dict_t:
                filtered_s.append((i,char))  #if find char in dict_t, append to filtered_s,indicateing char in which position
        
        lp = 0
        rp = 0  #two pointers
        
        formed = 0
        window_counts = {} #hashmap, {'char': # of occurence}
        
        ans = float("inf"),None,None
        
        while rp < len(filtered_s):
            character = filtered_s[rp][1]   #get char
            window_counts[character] = window_counts.get(character,0) + 1  #if no value, return 0 ,otherwise return char's value +1; +1 这次出现，value是之前的出现次数
            
            if window_counts[character] == dict_t[character]:  #如果自己数的value = t 的value，则 formed 这个计数 +1
                formed +=1
            
            while lp <= rp and formed == required:
                character = filtered_s[lp][1]
                end = filtered_s[rp][0]
                start = filtered_s[lp][0]
                
                if end - start + 1 < ans[0]:
                    ans = (end - start +1, start, end)
                
                window_counts[character] -=1
                
                if window_counts[character] < dict_t[character]:
                    formed -=1
                
                lp +=1
            rp +=1
        
        return "" if ans[0] == float("inf") else s[ans[1]:ans[2]+1]
        
```

```python
#Merged Sorted Array

# You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

# Merge nums1 and nums2 into a single array sorted in non-decreasing order.

# Example 1:
# Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
# Output: [1,2,2,3,5,6]

# Example 2:
# Input: nums1 = [1], m = 1, nums2 = [], n = 0
# Output: [1]

# Example 3:
# Input: nums1 = [0], m = 0, nums2 = [1], n = 1
# Output: [1]


class Solution(object):
    def merge(self,num1,m,num2,n):
        #指向从 0 到 n的位置
        for i in range(n):
            #然后将 num2对应位置的数，写入num1 m位置的后面
            num1[i+m] = num2[i]
        #进行排序，默认是升序
        num1.sort()
```

```python
# Valid Palindrome

# A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, 
# it reads the same forward and backward. Alphanumeric characters include letters and numbers.

# Given a string s, return true if it is a palindrome, or false otherwise.

# Example 1:
# Input: s = "A man, a plan, a canal: Panama"
# Output: true

# Example 2:
# Input: s = "race a car"
# Output: false



class Solution:
    def isPalindrome(self, s: str) -> bool:
        filtered_s = filter(lambda x: x.isalnum(),s)
        lower_case = map(lambda x: x.lower(),filtered_s)
        
        strings = list(lower_case)
        reverse_str = strings[::-1]
        
        return strings == reverse_str
```

```python
# Read N Characters Given Read4


# """
# The read4 API is already defined for you.

#     @param buf4, a list of characters
#     @return an integer
#     def read4(buf4):

# # Below is an example of how the read4 API can be called.
# file = File("abcdefghijk") # File is "abcdefghijk", initially file pointer (fp) points to 'a'
# buf4 = [' '] * 4 # Create buffer with enough space to store characters
# read4(buf4) # read4 returns 4. Now buf = ['a','b','c','d'], fp points to 'e'
# read4(buf4) # read4 returns 4. Now buf = ['e','f','g','h'], fp points to 'i'
# read4(buf4) # read4 returns 3. Now buf = ['i','j','k',...], fp points to end of file
# """

class Solution:
    def read(self, buf, n):
        """
        :type buf: Destination buffer (List[str])
        :type n: Number of characters to read (int)
        :rtype: The number of actual characters read (int)
        """
        copied_chars = 0
        read_chars = 4
        buf4 = [''] * 4
        
        while copied_chars < n and read_chars == 4:
            read_chars = read4(buf4)
            
            for i in range(read_chars):
                if copied_chars == n:
                    return copied_chars
                buf[copied_chars] = buf4[i]
                copied_chars +=1
                
```

```python
# Read N Characters Given Read4 II - Call multiple times

# Example 1:
# Given buf = "abc"
# read("abc", 1) // returns "a" # Should return 1
# read("abc", 2); // returns "bc" # Should return 2
# read("abc", 1); // returns "" # Should return 0
# Example 2:
# Given buf = "abc"
# read("abc", 4) // returns "abc" # Should return 3
# read("abc", 1); // returns "" # Should return 0

class Solution(object):
    def __init__(self):
        self.buff4Ptr = 0 # pointer
        self.buff4Cnt = 0 # count
        self.buff4 = [" "] * 4

    """
    @param {str[]} buf: Destination buffer
    @param {int} n: Maximum number of characters to read
    @return {str} The number of characters read
    """
    def read(self, buf, n):
        ptr = 0
        while ptr < n:
            if self.buff4Ptr == 0:
                self.buff4Cnt = read4(self.buff4)
            if self.buff4Cnt == 0:
                break
            
            while ptr < n and self.buff4Ptr < self.buff4Cnt:
                buf[ptr] = self.buff4[self.buff4Ptr]
                ptr += 1
                self.buff4Ptr += 1

            if self.buff4Ptr >= self.buff4Cnt:
                self.buff4Ptr = 0

        return ptr
```
