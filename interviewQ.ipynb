```python
# btc <--> eur
# usd <--> eth
# eth <--> eur

# "BTC-USD" { "ask": 1000, "bid": 900 }
# ask: 1 USD = 1/1000 BTC
# bid: 1 BTC = 900 USD

# "BTC-USD" { "ask": 1000, "bid": 900 }
# "BTC-EUR" { "ask": 1200, "bid": 1150 }
# "ETH-USD" { "ask": 200, "bid": 180 }
# "ETH-EUR" { "ask": 220, "bid": 210 }

# base_currency USD
# amount 1
# quote_currency EUR

from collections import defaultdict

class Exchange:
    def __init__(self,ask,bid):
        self.ask = ask
        self.bid = bid

class BestExchange:
    def __init__(self, trans):
        self.trans = trans
        self.neighs = defaultdict(list)
        for exchange in trans:
            temp_cur_a, temp_cur_b = exchange.split("-")
            exchange_rate = trans[exchange]
            self.neighs[temp_cur_a].append((temp_cur_b,exchange_rate.bid))
            self.neighs[temp_cur_b].append((temp_cur_a,1/ exchange_rate.ask))
        self.best_solution = None
        #print(self.neighs)

   
    def find_best_exchange(self,base_currency,quote_currency, amount,visited_currency):
        best_amount = None
        visited_currency.append(base_currency)
        for (target_currency, exchange_rate) in self.neighs[base_currency]:
            if target_currency in visited_currency:
                continue
            new_amount = amount * exchange_rate
            best_amount = new_amount if not best_amount else max(new_amount,best_amount)
            if target_currency != quote_currency:
                new_ans = self.find_best_exchange(target_currency,quote_currency, new_amount,visited_currency)
                if best_amount and new_ans:
                    best_amount = max(best_amount, new_ans)
                elif not best_amount:
                    best_amount = new_ans
               
        return best_amount


trans = {"BTC-USD":Exchange(1000,900), "BTC-EURP":Exchange(1200,1150),
"ETH-USD":Exchange(200,180), "ETH-EURP":Exchange(220,210)}
best_change = BestExchange(trans)
print(best_change.find_best_exchange("USD","EUR", 1, []))
```
